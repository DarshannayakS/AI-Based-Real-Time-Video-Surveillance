```python
import os
import cv2
import torch
import random
import numpy as np
import torch.nn as nn
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
from tqdm import tqdm
from torch.utils.data import Dataset, DataLoader, random_split
from torchvision.models.video import r3d_18
from sklearn.metrics import (
    confusion_matrix,
    f1_score,
    precision_score,
    recall_score,
    ConfusionMatrixDisplay,
)


device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)


IMG_SIZE    = 96
CLIP_LEN    = 8
BATCH_SIZE  = 4
EPOCHS      = 15


PATIENCE    = 5  

DATA_PATH = "/home/student/capstone_project/CCTV_camera/dataset"


classes = {"normal": 0, "anomaly": 1}

class VideoDataset(Dataset):
    def __init__(self, root):
        self.samples = []

        for cls in classes:
            folder = os.path.join(root, cls)
            for file in os.listdir(folder):
                if file.endswith((".mp4", ".avi", ".webm")):
                    self.samples.append(
                        (os.path.join(folder, file), classes[cls])
                    )

        print("Total videos:", len(self.samples))

    def load_video(self, path):
        cap = cv2.VideoCapture(path)
        frames = []

        while True:
            ret, frame = cap.read()
            if not ret:
                break

            frame = cv2.resize(frame, (IMG_SIZE, IMG_SIZE))
            frame = frame[:, :, ::-1]
            frames.append(frame)

        cap.release()

        if len(frames) < CLIP_LEN:
            frames = frames * (CLIP_LEN // len(frames) + 1)

        start = random.randint(0, len(frames) - CLIP_LEN)
        clip = frames[start : start + CLIP_LEN]

        clip = np.array(clip).transpose((3, 0, 1, 2))
        clip = torch.tensor(clip, dtype=torch.float32) / 255.0

        return clip

    def __len__(self):
        return len(self.samples)

    def __getitem__(self, idx):
        path, label = self.samples[idx]
        return self.load_video(path), label



dataset = VideoDataset(DATA_PATH)

train_size = int(0.8 * len(dataset))
test_size  = len(dataset) - train_size

train_dataset, test_dataset = random_split(dataset, [train_size, test_size])

train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)
test_loader  = DataLoader(test_dataset,  batch_size=BATCH_SIZE, shuffle=False)

print(f"Train samples : {train_size}")
print(f"Test  samples : {test_size}")


model = r3d_18(weights="DEFAULT")


for param in model.layer1.parameters():
    param.requires_grad = False

for param in model.layer2.parameters():
    param.requires_grad = False

model.fc = nn.Linear(model.fc.in_features, 2)
model    = model.to(device)


criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=1e-4)


train_losses = []
test_losses  = []
train_accs   = []
test_accs    = []

f1_scores    = []
precisions   = []
recalls      = []

os.makedirs("checkpoints",        exist_ok=True)
os.makedirs("confusion_matrices", exist_ok=True)

best_acc          = 0.0
epochs_no_improve = 0   
class_names       = ["normal", "anomaly"]


def save_confusion_matrix(all_labels, all_preds, epoch):
    cm  = confusion_matrix(all_labels, all_preds)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=class_names)

    fig, ax = plt.subplots(figsize=(5, 4))
    disp.plot(ax=ax, colorbar=False, cmap="Blues")
    ax.set_title(f"Confusion Matrix — Epoch {epoch}")
    plt.tight_layout()
    plt.savefig(f"confusion_matrices/epoch_{epoch}.png", dpi=100)
    plt.close(fig)

for epoch in range(1, EPOCHS + 1):

    print(f"\n===== Epoch {epoch}/{EPOCHS} =====")

    model.train()
    total_loss = 0
    correct    = 0
    total      = 0

    for videos, labels in tqdm(train_loader, desc="Training"):
        videos = videos.to(device)
        labels = labels.to(device)

        outputs = model(videos)
        loss    = criterion(outputs, labels)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        total_loss += loss.item()

        _, pred  = outputs.max(1)
        total   += labels.size(0)
        correct += pred.eq(labels).sum().item()

    train_loss = total_loss / len(train_loader)
    train_acc  = 100 * correct / total

    train_losses.append(train_loss)
    train_accs.append(train_acc)

    print(f"Train Loss : {train_loss:.4f}")
    print(f"Train Acc  : {train_acc:.2f}%")

    model.eval()
    total_loss = 0
    correct    = 0
    total      = 0

    all_labels = []
    all_preds  = []

    with torch.no_grad():
        for videos, labels in tqdm(test_loader, desc="Testing"):
            videos = videos.to(device)
            labels = labels.to(device)

            outputs = model(videos)
            loss    = criterion(outputs, labels)

            total_loss += loss.item()

            _, pred  = outputs.max(1)
            total   += labels.size(0)
            correct += pred.eq(labels).sum().item()

            all_labels.extend(labels.cpu().numpy())
            all_preds.extend(pred.cpu().numpy())

    test_loss = total_loss / len(test_loader)
    test_acc  = 100 * correct / total

    test_losses.append(test_loss)
    test_accs.append(test_acc)


    f1        = f1_score(all_labels, all_preds, average="weighted", zero_division=0)
    precision = precision_score(all_labels, all_preds, average="weighted", zero_division=0)
    recall    = recall_score(all_labels, all_preds, average="weighted", zero_division=0)

    f1_scores.append(f1)
    precisions.append(precision)
    recalls.append(recall)

    print(f"Test  Loss : {test_loss:.4f}")
    print(f"Test  Acc  : {test_acc:.2f}%")
    print(f"F1 Score   : {f1:.4f}")
    print(f"Precision  : {precision:.4f}")
    print(f"Recall     : {recall:.4f}")

    save_confusion_matrix(all_labels, all_preds, epoch)
    print(f"Confusion matrix saved → confusion_matrices/epoch_{epoch}.png")

    torch.save(model.state_dict(), f"checkpoints/epoch_{epoch}.pth")
    print(f"Checkpoint saved → checkpoints/epoch_{epoch}.pth")


    if test_acc > best_acc:
        best_acc          = test_acc
        epochs_no_improve = 0
        torch.save(model.state_dict(), "best_model.pth")
        print("Best model saved!")
    else:
        epochs_no_improve += 1
        print(f"No improvement for {epochs_no_improve}/{PATIENCE} epochs.")

        if epochs_no_improve >= PATIENCE:
            print(f"\n Early stopping triggered after epoch {epoch}.")
            break



epochs_ran = range(1, len(train_accs) + 1)

fig = plt.figure(figsize=(14, 10))
gs  = gridspec.GridSpec(2, 2, figure=fig)


ax1 = fig.add_subplot(gs[0, 0])
ax1.plot(epochs_ran, train_accs, marker="o", label="Train Acc")
ax1.plot(epochs_ran, test_accs,  marker="o", label="Test Acc")
ax1.set_title("Accuracy per Epoch")
ax1.set_xlabel("Epoch")
ax1.set_ylabel("Accuracy (%)")
ax1.legend()
ax1.grid(True)


ax2 = fig.add_subplot(gs[0, 1])
ax2.plot(epochs_ran, train_losses, marker="o", label="Train Loss")
ax2.plot(epochs_ran, test_losses,  marker="o", label="Test Loss")
ax2.set_title("Loss per Epoch")
ax2.set_xlabel("Epoch")
ax2.set_ylabel("Loss")
ax2.legend()
ax2.grid(True)


ax3 = fig.add_subplot(gs[1, :])
ax3.plot(epochs_ran, f1_scores,  marker="o", label="F1 Score")
ax3.plot(epochs_ran, precisions, marker="s", label="Precision")
ax3.plot(epochs_ran, recalls,    marker="^", label="Recall")
ax3.set_title("F1 Score, Precision & Recall per Epoch")
ax3.set_xlabel("Epoch")
ax3.set_ylabel("Score")
ax3.legend()
ax3.grid(True)

plt.suptitle("Training Performance", fontsize=16, fontweight="bold")
plt.tight_layout()
plt.savefig("training_graph.png", dpi=120)
plt.show()

print("\n Training Completed!")
print(f"Best Test Accuracy : {best_acc:.2f}%")
```

```python
import cv2
import torch
import numpy as np
import torch.nn as nn
import smtplib
import os
import time
import threading
import subprocess
from datetime import datetime
from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText
from email.mime.base import MIMEBase
from email import encoders
from torchvision.models.video import r3d_18


VIDEO_PATH = "/home/student/Downloads/anomaly_video/clipsa_086.mp4"
MODEL_PATH = "/home/student/capstone_project/checkpoints/epoch_10.pth"

SENDER_EMAIL = "2022darshan.s@vidyashilp.edu.in"
APP_PASSWORD = "ymlw nkra qsrm yvex"
OWNER_EMAIL = "sdarshannayak@gmail.com"

SAVE_DIR = "/home/student/capstone_project/anomaly_clips"
os.makedirs(SAVE_DIR, exist_ok=True)

IMG_SIZE = 96
CLIP_LEN = 8
EMAIL_COOLDOWN = 60
RECORD_SECONDS = 4

classes = ["normal", "anomaly"]

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)


model = r3d_18(weights=None)
model.fc = nn.Linear(model.fc.in_features, 2)
model.load_state_dict(torch.load(MODEL_PATH, map_location=device))
model = model.to(device)
model.eval()

print(f" Model loaded from: {MODEL_PATH}")


def convert_to_mp4(input_path):
    output_path = input_path.replace(".avi", ".mp4")

    command = [
        "ffmpeg",
        "-y",
        "-i", input_path,
        "-vcodec", "libx264",
        "-acodec", "aac",
        "-movflags", "+faststart",
        output_path
    ]

    subprocess.run(command, stdout=subprocess.PIPE, stderr=subprocess.PIPE)

    if os.path.exists(input_path):
        os.remove(input_path)

    return output_path


def send_email(video_path, frame_no, label, conf):
    try:
        print("📧 Sending email...")

        msg = MIMEMultipart()
        msg["From"] = SENDER_EMAIL
        msg["To"] = OWNER_EMAIL
        msg["Subject"] = " Anomaly Detected (VIDEO)"

        body = f"""
Frame: {frame_no}
Prediction: {label}
Confidence: {conf:.2f}
Time: {datetime.now()}
"""
        msg.attach(MIMEText(body, "plain"))

        with open(video_path, "rb") as f:
            part = MIMEBase("application", "octet-stream")
            part.set_payload(f.read())
            encoders.encode_base64(part)
            part.add_header(
                "Content-Disposition",
                f"attachment; filename={os.path.basename(video_path)}"
            )
            msg.attach(part)

        with smtplib.SMTP_SSL("smtp.gmail.com", 465) as server:
            server.login(SENDER_EMAIL, APP_PASSWORD)
            server.send_message(msg)

        print("EMAIL SENT SUCCESSFULLY")

    except Exception as e:
        print("Email Error:", e)


class ClipRecorder:
    def __init__(self, fps, width, height):
        self.fps = max(fps, 1)
        self.width = width
        self.height = height
        self.writer = None
        self.frames_left = 0
        self.path = None
        self.recording = False

    def start(self):
        if self.recording:
            return

        timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
        self.path = os.path.join(SAVE_DIR, f"anomaly_{timestamp}.avi")

        self.writer = cv2.VideoWriter(
            self.path,
            cv2.VideoWriter_fourcc(*"XVID"),
            self.fps,
            (self.width, self.height)
        )

        self.frames_left = int(self.fps * RECORD_SECONDS)
        self.recording = True
        print("Recording anomaly clip...")

    def write(self, frame):
        if self.recording and self.frames_left > 0:
            self.writer.write(frame)
            self.frames_left -= 1

        if self.frames_left <= 0 and self.recording:
            self.stop()

    def stop(self):
        if self.writer:
            self.writer.release()
        self.recording = False
        print("Clip saved:", self.path)

def preprocess_clip(frames):
    clip = np.array(frames).transpose((3, 0, 1, 2))
    clip = torch.tensor(clip, dtype=torch.float32) / 255.0
    return clip.unsqueeze(0).to(device)


def run():
    cap = cv2.VideoCapture(VIDEO_PATH)

    if not cap.isOpened():
        print("Cannot open video")
        return

    fps = cap.get(cv2.CAP_PROP_FPS) or 25
    width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

    recorder = ClipRecorder(fps, width, height)

    buffer = []
    predictions = []
    last_email_time = 0
    frame_id = 0


    normal_count = 0
    anomaly_count = 0

    while True:
        ret, frame = cap.read()
        if not ret:
            break

        frame_id += 1

        resized = cv2.resize(frame, (IMG_SIZE, IMG_SIZE))
        rgb = resized[:, :, ::-1]
        buffer.append(rgb)

        if len(buffer) == CLIP_LEN:
            clip = preprocess_clip(buffer)

            with torch.no_grad():
                output = model(clip)
                probs = torch.softmax(output, dim=1)
                pred = torch.argmax(probs, dim=1).item()
                conf = probs[0][pred].item()

            predictions.append(pred)
            if len(predictions) > 10:
                predictions.pop(0)

            final_pred = max(set(predictions), key=predictions.count)


            if final_pred == 0:
                normal_count += 1
            else:
                anomaly_count += 1

 
            print(f"Frame {frame_id}: {classes[final_pred]} ({conf:.2f})")

            buffer.pop(0)

            if final_pred == 1:
                recorder.start()

                current_time = time.time()
                if current_time - last_email_time > EMAIL_COOLDOWN:
                    last_email_time = current_time

                    def send_async(path):
                        time.sleep(RECORD_SECONDS + 1)

                        if os.path.exists(path):
                            print(" Converting to MP4...")
                            mp4_path = convert_to_mp4(path)

                            print(" Sending anomaly clip...")
                            send_email(mp4_path, frame_id, "ANOMALY", conf)

                    threading.Thread(
                        target=send_async,
                        args=(recorder.path,),
                        daemon=True
                    ).start()

        recorder.write(frame)

        cv2.imshow("Surveillance", frame)

        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    cap.release()
    cv2.destroyAllWindows()


    print("\n Finished Processing")

    total = normal_count + anomaly_count

    if total == 0:
        final_result = "UNKNOWN"
        anomaly_ratio = 0
    else:
        anomaly_ratio = anomaly_count / total

        if anomaly_ratio > 0.6:
            final_result = "ANOMALY"
        else:
            final_result = "NORMAL"

    print("\n-------FINAL REPORT --------")
    print(f"Normal clips   : {normal_count}")
    print(f"Anomaly clips  : {anomaly_count}")
    print(f"Anomaly ratio  : {anomaly_ratio:.2f}")
    print(f"FINAL RESULT   : {final_result}")
    print("\n")


run()
```
